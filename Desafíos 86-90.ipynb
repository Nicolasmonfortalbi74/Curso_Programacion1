```python
### Desaf√≠o 86: Crear un archivo de pr√©stamos de libros
Desarrolla un programa que cree un archivo prestamos.txt y permita al usuario agregar el registro de un pr√©stamo.
El registro debe incluir el nombre del libro, el nombre del prestatario y la fecha del pr√©stamo. 
Aseg√∫rate de no sobrescribir el archivo cada vez que se agrega un nuevo pr√©stamo.
```

```python
from datetime import datetime

while True:
    print("\n1. Agregar pr√©stamo")
    print("2. Mostrar pr√©stamos")
    print("3. Salir")
    opcion = input("Elige una opci√≥n: ")

    if opcion == "1":
        libro = input("Nombre del libro: ")
        prestatario = input("Nombre del prestatario: ")
        fecha = input("Fecha del pr√©stamo (YYYY-MM-DD): ")

        try:
            datetime.strptime(fecha, "%Y-%m-%d")  # valida el formato de la fecha
            with open("prestamos.txt", "a", encoding="utf-8") as archivo:
                archivo.write(f"{libro} | {prestatario} | {fecha}\n")
            print("Pr√©stamo registrado correctamente.")
        except ValueError:
            print("Fecha inv√°lida. Usa el formato YYYY-MM-DD.")

    elif opcion == "2":
        try:
            with open("prestamos.txt", "r", encoding="utf-8") as archivo:
                contenido = archivo.read()
                if contenido:
                    print("\n--- Lista de pr√©stamos ---")
                    print(contenido)
                else:
                    print("No hay pr√©stamos registrados.")
        except FileNotFoundError:
            print("El archivo a√∫n no existe. No hay pr√©stamos registrados.")

    elif opcion == "3":
        print("Saliendo del programa...")
        break
    else:
        print("Opci√≥n no v√°lida.")
```

Bucle principal (while True)
El programa muestra un peque√±o men√∫ para elegir entre:

1 ‚Üí Agregar un pr√©stamo

2 ‚Üí Mostrar los pr√©stamos

3 ‚Üí Salir del programa

Agregar pr√©stamo (opci√≥n 1)

Se piden tres datos: nombre del libro, del prestatario y la fecha.

Se valida la fecha con datetime.strptime() para asegurar que est√© en formato YYYY-MM-DD.

Se abre el archivo prestamos.txt en modo ‚Äúa‚Äù (append) para agregar al final sin borrar lo anterior.

Se escribe una l√≠nea con los datos y un salto de l√≠nea.

Mostrar pr√©stamos (opci√≥n 2)

Se intenta abrir el archivo en modo lectura ("r").

Si no existe, se muestra un mensaje.

Si existe y tiene contenido, se imprime tal cual.

Salir (opci√≥n 3)
Rompe el bucle con break y finaliza el programa.

### Desaf√≠o 87: Buscar un libro por autor en un archivo de texto
Dado un archivo libros.txt que contiene una lista de libros y sus autores, implementa una funci√≥n que busque todos los libros escritos por un autor espec√≠fico y los muestre. Si el autor no tiene libros en la lista, debe mostrar un mensaje indicando que no hay coincidencias.

```python
def buscar_libros_por_autor(nombre_autor):
    try:
        with open("libros.txt", "r", encoding="utf-8") as archivo:
            lineas = archivo.readlines()

        encontrados = []

        for linea in lineas:
            # Dividir cada l√≠nea en t√≠tulo y autor usando "|"
            partes = linea.strip().split("|")
            if len(partes) == 2:
                titulo = partes[0].strip()
                autor = partes[1].strip()

                # Comparar nombres sin importar may√∫sculas o min√∫sculas
                if autor.lower() == nombre_autor.lower():
                    encontrados.append(titulo)

        if encontrados:
            print("\nLibros del autor", nombre_autor + ":")
            for libro in encontrados:
                print("-", libro)
        else:
            print("No se encontraron libros de ese autor.")

    except FileNotFoundError:
        print("El archivo libros.txt no existe. Aseg√∫rate de crearlo primero.")


# Programa principal
autor = input("Ingrese el nombre del autor que desea buscar: ")
buscar_libros_por_autor(autor)
```

### Desaf√≠o 90: Contar las palabras m√°s comunes en un archivo de texto Desarrolla una funci√≥n que lea un archivo de texto grande, como libros.txt, y cuente cu√°ntas veces aparece cada palabra. Luego, muestra las 5 palabras m√°s comunes y cu√°ntas veces aparecen.

```python
from collections import Counter
import re

def palabras_mas_comunes(nombre_archivo, top=5):
    try:
        # Leer todo el contenido del archivo
        with open(nombre_archivo, "r", encoding="utf-8") as f:
            texto = f.read().lower()  # pasamos a min√∫sculas para un conteo uniforme

        # Usar una expresi√≥n regular para extraer solo palabras (sin puntuaci√≥n)
        palabras = re.findall(r"\b[a-z√°√©√≠√≥√∫√º√±]+(?:'[a-z√°√©√≠√≥√∫√º√±]+)?\b", texto, re.IGNORECASE)

        # Contar la frecuencia de cada palabra
        contador = Counter(palabras)

        # Obtener las 5 m√°s comunes
        mas_comunes = contador.most_common(top)

        # Mostrar los resultados
        print(f"\nüìò Las {top} palabras m√°s comunes en '{nombre_archivo}' son:\n")
        for palabra, cantidad in mas_comunes:
            print(f"‚Ä¢ {palabra}: {cantidad} veces")

    except FileNotFoundError:
        print(f"‚ùå El archivo '{nombre_archivo}' no existe.")
    except Exception as e:
        print("Ocurri√≥ un error:", e)


# Ejemplo de uso
palabras_mas_comunes("libros.txt")
```
