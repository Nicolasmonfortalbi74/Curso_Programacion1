```python
### Desaf√≠o 81: Recorrido de estudiantes por niveles
Dado un √°rbol que representa los grupos de estudiantes en una escuela, implementa un recorrido por niveles para mostrar los estudiantes de cada grupo, comenzando por el nivel m√°s alto (ej. grado 12)
y descendiendo hasta el nivel m√°s bajo (ej. grado 1).
Cada nodo del √°rbol representa un grado y sus estudiantes.
```

```python
Grado 12
 ‚îú‚îÄ‚îÄ Grado 11
 ‚îÇ    ‚îú‚îÄ‚îÄ Grado 10
 ‚îÇ    ‚îî‚îÄ‚îÄ Grado 9
 ‚îî‚îÄ‚îÄ Grado 8
      ‚îú‚îÄ‚îÄ Grado 7
      ‚îî‚îÄ‚îÄ Grado 6
```

```python
from collections import deque

# Clase para representar cada nodo del √°rbol
class NodoGrado:
    def __init__(self, grado, estudiantes):
        self.grado = grado
        self.estudiantes = estudiantes
        self.hijos = []  # subgrados o divisiones del grado

# Funci√≥n para recorrer el √°rbol por niveles
def recorrido_por_niveles(raiz):
    if not raiz:
        return
    
    cola = deque([raiz])  # usamos una cola para el recorrido
    
    while cola:
        nodo_actual = cola.popleft()  # sacamos el primer nodo de la cola
        print(f"üéì {nodo_actual.grado}: {', '.join(nodo_actual.estudiantes)}")
        
        # agregamos los hijos a la cola
        for hijo in nodo_actual.hijos:
            cola.append(hijo)

# --- Ejemplo de uso ---

# Crear los nodos (grados)
grado12 = NodoGrado("Grado 12", ["Ana", "Luis"])
grado11 = NodoGrado("Grado 11", ["Sof√≠a", "Carlos"])
grado10 = NodoGrado("Grado 10", ["Mart√≠n", "Laura"])
grado9  = NodoGrado("Grado 9",  ["Juli√°n", "Valentina"])
grado8  = NodoGrado("Grado 8",  ["Diego", "Camila"])
grado7  = NodoGrado("Grado 7",  ["Tom√°s", "Luc√≠a"])
grado6  = NodoGrado("Grado 6",  ["Mateo", "Paula"])

# Construir el √°rbol (niveles jer√°rquicos)
grado12.hijos = [grado11, grado8]
grado11.hijos = [grado10, grado9]
grado8.hijos = [grado7, grado6]

# Recorrido
print("üè´ Recorrido de estudiantes por niveles:")
recorrido_por_niveles(grado12)
```

Creamos una cola con el nodo ra√≠z (Grado 12).

Mientras la cola no est√© vac√≠a, sacamos un nodo y lo mostramos.

Agregamos sus hijos al final de la cola.

El orden en que los sacamos de la cola asegura que recorramos por niveles.

```python
### Desaf√≠o 82: Implementar b√∫squeda secuencial en una tabla de calificaciones
Tienes una tabla de calificaciones representada como una matriz, 
donde cada fila contiene las calificaciones de un estudiante en distintas materias. 
Implementa una funci√≥n que busque una calificaci√≥n espec√≠fica 
en toda la matriz y devuelva el estudiante y la materia en la que se encuentra.
```

```python
def buscar_calificacion(matriz, estudiantes, materias, calificacion_buscada):
    # Recorremos las filas (estudiantes)
    for i in range(len(matriz)):
        # Recorremos las columnas (materias)
        for j in range(len(matriz[i])):
            if matriz[i][j] == calificacion_buscada:
                return f"‚úÖ Calificaci√≥n {calificacion_buscada} encontrada: {estudiantes[i]} en {materias[j]}"
    
    # Si no se encuentra
    return f"‚ùå Calificaci√≥n {calificacion_buscada} no encontrada en la tabla."

# --- Ejemplo de uso ---

estudiantes = ["Ana", "Luis", "Sof√≠a"]
materias = ["Matem√°tica", "Historia", "Ingl√©s"]

calificaciones = [
    [9, 8, 10],   # Ana
    [6, 7, 9],    # Luis
    [10, 9, 8]    # Sof√≠a
]

# Buscar una nota espec√≠fica
print(buscar_calificacion(calificaciones, estudiantes, materias, 7))
print(buscar_calificacion(calificaciones, estudiantes, materias, 5))
```

Recorremos la matriz con dos bucles for:

El primero (√≠ndice i) recorre los estudiantes (filas).

El segundo (√≠ndice j) recorre las materias (columnas).

En cada posici√≥n [i][j], comparamos la calificaci√≥n con la buscada.

Si coincide, retornamos los nombres del estudiante y la materia correspondientes.

Si no se encuentra, devolvemos un mensaje de error.

### Desaf√≠o 83: Optimizar la b√∫squeda en una lista ordenada de estudiantes
Tienes una lista ordenada alfab√©ticamente con los nombres de los estudiantes de una clase. Implementa una funci√≥n que realice una b√∫squeda binaria para encontrar un estudiante espec√≠fico en la lista. Si el estudiante no est√°, la funci√≥n debe mostrar un mensaje adecuado.

```python
def buscar_estudiante(lista, nombre_buscado):
    inicio = 0
    fin = len(lista) - 1

    while inicio <= fin:
        medio = (inicio + fin) // 2
        actual = lista[medio]

        if actual == nombre_buscado:
            return f"‚úÖ {nombre_buscado} encontrado en la posici√≥n {medio}."
        elif nombre_buscado < actual:
            fin = medio - 1
        else:
            inicio = medio + 1

    return f"‚ùå {nombre_buscado} no se encuentra en la lista."

# --- Ejemplo de uso ---
estudiantes = ["Ana", "Carlos", "Javier", "Luis", "Sof√≠a"]

print(buscar_estudiante(estudiantes, "Luis"))
print(buscar_estudiante(estudiantes, "Mar√≠a"))
```

inicio y fin marcan los extremos de la lista.

Calculamos medio y comparamos:

Si el nombre coincide ‚Üí lo encontramos.

Si el nombre buscado es alfab√©ticamente menor, buscamos a la izquierda.

Si es mayor, buscamos a la derecha.

El proceso se repite hasta que inicio > fin.

### Desaf√≠o 84: Ordenar estudiantes por promedio de calificaciones
Tienes una lista de estudiantes y su promedio de calificaciones. Implementa un algoritmo que ordene a los estudiantes de acuerdo con su promedio utilizando el algoritmo de ordenamiento por selecci√≥n. Al final, el estudiante con el promedio m√°s alto debe estar en primer lugar.

```python
def ordenar_por_promedio(estudiantes):
    n = len(estudiantes)
    
    # Recorremos toda la lista
    for i in range(n - 1):
        # Suponemos que el m√°ximo est√° en la posici√≥n i
        indice_max = i
        
        # Buscamos el mayor promedio en el resto de la lista
        for j in range(i + 1, n):
            if estudiantes[j][1] > estudiantes[indice_max][1]:
                indice_max = j
        
        # Intercambiamos si encontramos un promedio mayor
        if indice_max != i:
            estudiantes[i], estudiantes[indice_max] = estudiantes[indice_max], estudiantes[i]
    
    return estudiantes

# --- Ejemplo de uso ---
estudiantes = [
    ("Ana", 8.5),
    ("Luis", 6.7),
    ("Sof√≠a", 9.2),
    ("Carlos", 7.8)
]

ordenados = ordenar_por_promedio(estudiantes)

print("üè´ Estudiantes ordenados por promedio (de mayor a menor):")
for nombre, promedio in ordenados:
    print(f"{nombre}: {promedio}")
```

Recorremos la lista desde i = 0 hasta n - 1.

Para cada posici√≥n i, buscamos el estudiante con el mayor promedio en el resto de la lista.

Intercambiamos el estudiante actual con el de mayor promedio encontrado.

Al terminar, la lista est√° ordenada de mayor a menor.

### Desaf√≠o 85: Crear un √°rbol de clasificaci√≥n de estudiantes por rendimiento
Dado un conjunto de estudiantes y sus promedios, implementa una funci√≥n que cree un √°rbol binario de b√∫squeda en el que los nodos representan los promedios de los estudiantes. Luego, implementa una funci√≥n que recorra el √°rbol en inorden para mostrar los estudiantes en orden ascendente de rendimiento acad√©mico.

```python
# Nodo del BST: cada nodo contiene:
# - clave: promedio (float)
# - estudiantes: lista de nombres que comparten ese promedio
# - izquierdo / derecho: hijos del BST
class NodoPromedio:
    def __init__(self, promedio, nombre):
        self.promedio = promedio
        self.estudiantes = [nombre]  # lista para manejar duplicados
        self.izquierdo = None
        self.derecho = None

# Inserta un estudiante en el BST seg√∫n su promedio
def insertar(raiz, promedio, nombre):
    if raiz is None:
        return NodoPromedio(promedio, nombre)
    # Si el promedio ya existe, agregamos el nombre al nodo
    if promedio == raiz.promedio:
        raiz.estudiantes.append(nombre)
    elif promedio < raiz.promedio:
        raiz.izquierdo = insertar(raiz.izquierdo, promedio, nombre)
    else:
        raiz.derecho = insertar(raiz.derecho, promedio, nombre)
    return raiz

# Recorrido inorden: devuelve lista de tuplas (promedio, [nombres]) ordenadas ascendentemente
def recorrido_inorden(raiz, resultado=None):
    if resultado is None:
        resultado = []
    if raiz is None:
        return resultado
    recorrido_inorden(raiz.izquierdo, resultado)
    resultado.append((raiz.promedio, raiz.estudiantes.copy()))
    recorrido_inorden(raiz.derecho, resultado)
    return resultado

# Funci√≥n auxiliar: construye el √°rbol a partir de una lista de (nombre, promedio)
def construir_arbol(lista_estudiantes):
    raiz = None
    for nombre, promedio in lista_estudiantes:
        raiz = insertar(raiz, promedio, nombre)
    return raiz

# --- Ejemplo de uso ---
if __name__ == "__main__":
    datos = [
        ("Ana", 8.5),
        ("Luis", 6.7),
        ("Sof√≠a", 9.2),
        ("Carlos", 7.8),
        ("Mar√≠a", 8.5),   # mismo promedio que Ana
        ("Pedro", 6.7),   # mismo promedio que Luis
        ("Luc√≠a", 10.0)
    ]

    raiz = construir_arbol(datos)
    orden_ascendente = recorrido_inorden(raiz)

    print("üìã Estudiantes ordenados por promedio (ascendente):")
    for promedio, nombres in orden_ascendente:
        # mostramos todos los estudiantes que comparten el promedio
        print(f"{promedio}: {', '.join(nombres)}")
```

Cada nodo del √°rbol guarda un promedio y la lista de estudiantes con ese promedio.

Al insertar, si el promedio es menor, va a la izquierda; si es mayor, a la derecha; si es igual, se agrega al mismo nodo.

As√≠ se forma un √°rbol binario de b√∫squeda (BST) donde los promedios m√°s bajos quedan a la izquierda y los m√°s altos a la derecha.

El recorrido inorden (izquierda ‚Üí nodo ‚Üí derecha) recorre los nodos en orden ascendente de promedio.

El resultado final muestra los estudiantes del menor al mayor promedio.

 En resumen:
Construyes el √°rbol con los promedios y luego lo recorres en inorden para verlos ordenados de menor a mayor.
